### Structure and modules
Instamatic GUI is divided into three sectors. The entire left side is taken by the videostream frame, where a live view of the camera is shown. The top of this frame displays live view fps/interval and has some fields to change the size of the view, exposure time used for the live view, as well as brightness and display range used. On the bottom, the "Save frame" and "Save image" buttons will dump the current frame/frame into a png/tiff file in the current working directory.

The top right panel is reserved for the input-output (io) frame display. *Directory* allows to change the current working directory, which by default is `C:/instamatic/$date/`. *Sample name* and number determine the subdirectory where experimental data is stored. The number increments automatically when a new experiment is started, so that data is never overwritten. *Flatfield* can be determined here as well. *Open work directory* and *Open settings directory* buttons are mostly self-explanatory, while *Delete last experiment* as slightly misleading, as it only renames the folder for deletion: to avoid costly issues the final step must be always done by the user.

The bottom right panel stores variety of so-called "experiment" model frames designed to perform series of operations on the camera and microscope. Each panel reveals access to different kind of experiment design or functionality, has its own author, reference, requirements, and limitations. ### Control frame
Instamatic is a composite software consisting of microscope and camera servers, calibration, scripting, and experimental routines. The easiest way to make sure it handles the connection to TEM as intended is via the "Control" frame. In this frame, the user can change alpha angle and induce its wobbling, set stage speed setting, move the stage, and toggle some of the beam characteristics.

![Image of control frame](https://lh3.googleusercontent.com/sitesv/AA5AbUDwcAbtRgMBdsMmCdMNhOl2wLRPIRE0rgIME_TmIToHrIhr2nP5Y0vDJ68PLwnJqbJqSet8VnS1LeqNLXI4h5c3HMlm6a-4yLh7JNJBvv70o5aLQjzOrHd1XrDUIbfKbshAbPvAgQGqPmgeTOt62NGfXk_Dhj_C0sO7ZHFNBF0BiJ6e5EDdCMDQ4EQ3tBEY9_fMXGU-O1JvkJxJeEhbprOf6xyXSe5mJpNUUZM=w1280)

When starting work with Instamatic for the first time, it is a good idea to play a bit with the control frame to feel out how does the microscope react to the changes, which features do not work as expected, which are difficult to GUI. In particular, some of the features i.e. the magnification mode might not work properly depending on their implementation on the server. Nonetheless, this frame acts as a good place to confirm the connection works as intended.

### RED frame
In the simplest experimental scenario, when no tracking is required and stills in equal intervals is all that is needed, a simple RED experiment will do the trick. This experiment can be run using two different experimental frames, "RED" and "FastADT", the former being far simpler and raw one.

![Image of RED frame](https://lh3.googleusercontent.com/sitesv/AA5AbUD5Dg75BjhrG1vMkVWYXfMa8GMcEyiis6ZNl9zdFpbhEmvxjUtALxMma9WTK63uCm-dyhF6MT2jWbiW3Vh2a1ByCw2sIHa1RW5yR-7EJKGs4I6sSSfETmJcCSjqEfiQlj9xfXe8onrkt3sUfvhEMGiaOWaJvLpy8lcHnj5eC6Xer11b8ehVIUehzcyfx9TDF0USJ-kATy7spmw2TXGGLXhhQoRCGIohNm0kAio=w1280)

In the RED routine, the user can only set three parameters: exposure time, tilt range, and step size. Upon starting a RED experiment, the stage will march tilt range in positive direction, stopping and collecting a still diffraction image every step, finishing "range" degrees away from its start position. It offers no tracking nor complex control capabilities. This offers a simple way to test whether the program can record and save experiments correctly.

### cRED frame

The cRED frame offers the original Instamatic 3D ED experience. Designed for stages that feature smooth but irreproducible float, cRED frame can be used to collect continuous diffraction data where the crystal drift is corrected manually using beam shift. Diffraction data is collected and stage is rotated continuously, the former ends with the click of the button, and the latter automatically. For the best quality of data, both movements should be terminated at the same time.

![Image of cRED frame](https://lh3.googleusercontent.com/sitesv/AA5AbUB8lm0nfsFgv8py6P6U-iwFv676oVDATE9onAsd-4_tI5GLvj0zfyqPna720kdV4uu0_KsNU_zsQLeGhOy9zV2s3QIuca-Hammzikt_Pa9BmlFcUTDITNLf4hpiYjbr8TZODQSMRY4ghKH5vriZHQ39jyxdHaMRNSFaPqnPU7cc3tc6SGs9KfT7TfeSUa1yF1ZJD7gU5g_fnSn-Wml2pz-hh-bRDZnNylpjy3k=w1280)

The cRED frame has been originally designed for a JEOL microscope, where the stage can be rotated using dedicated foot pedals. It can be used without any tracking, but if one is desired, it has to be performed manually. Beam shift position can be then controlled using either the track ball or using control frame option to move the beam using Right Mouse Button. Both options require some experience, and the latter additionally relies on an up-to-date beamshift calibration file present in the current working directory (see `instamatic.calibrate_beamshift`).

### FastADT frame

The "FastADT" has been added in 2025 by the author as a functional replacement of the analogous protocol implemented as RATS: an iTEM plugin created by Yaşar Krysiak, Sergi Plana-Ruiz, and Lukáš Palatinus. It is capable of collecting RED, PED, and cRED data with a priori tracking. Compared to other single-crystal frames of Instamatic, it is robust and transferable but requires a stage with reproducible movement, as well as additional calibrations to run some of the cases.

![Image of FastADT frame](https://lh3.googleusercontent.com/sitesv/AA5AbUCc1a5eQJBPz9dGPi8d0a9OWWCr87OONRdEHkiHa-Ddc0TJAhVCRzlgjfU11Ma-BtkCFpwYb5Ktp0bRFT35tL_glzzYWtf_GbCA3Q_WKd-T0WpUGanqeaC6YaxORXqPd2LhrdB94siV1gDOzcRyZE5ZlW3ywyQfQVh4pV_TMWr8XUkv69M3b4ygQMXZqB5xp8IasIyV5ewTlc8yPSSKGPqewEWxlyuuQCNVWkc=w1280)

To start a "FastADT" experiment, select diffraction and tracking settings. Additionally, manually select and "store" the beam settings for taking "image", "tracking" crystal, and collecting "diffraction". These settings will be automatically recalled as needed during the experiment, and can be recalled manually using the "restore" buttons. Additionally, "beam blank" can be controlled using a dedicated button. Once everything is set, click "start" and follow the instructions on the screen (if any).

"Continuous" diffraction mode requires either a fitting `instamatic.calibrate_movie_delays` at target exposure to be complete or config.camera.dead_time to be set. "Manual" tracking algorithm requires `instamatic.calibrate_beamshift` in the current experiment directory, and this calibration will be run automatically with the current diffractions. As goes for all experiments, `config.calibrate` must also include the calibration constant for the current detector distance used in the diffraction mode.

### serialED and autocRED frames
Either of the last two currently publicly-available frames are designed to collect serial diffraction data. The first one, serialED, is actually one of the oldest additions to the code. It is designed to take a low-magnification image of the diffraction grid, identify crystals, and collect a single static diffraction frame from each. AutocRED, or SerialRED, the other module, additionally attempts automated collecting continuous rotation using in-situ crystal stracking. Unfortunately, both frames are relatively poorly documented and require rich calibration that is additionally often non-transferable. Due to these reasons, getting these two scientifically-excellent frames to work on new systems poses a significant challenge.

### Learning, advanced, about
These three remaining frames are not specifically designed to run experiment themselves, at least not by default. Learning may integrate with other routines and be used for neural network training; advanced allows for flat-field collection, running arbitrary scripts (click "Browse" to check them out), and auxiliary features for automated cRED. About hosts the most recent links and references to the program.
